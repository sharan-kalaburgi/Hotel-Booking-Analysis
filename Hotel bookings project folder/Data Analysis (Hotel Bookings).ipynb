# Data Analysis and visualizations

```python
canceled_per=df['is_canceled'].value_counts(normalize=True)
print(canceled_per)                #Gives the percentages of cancelled and not cancelled bookings from the is_cancelled column

plt.figure(figsize=(5,5))          #sets the figure size as mentioned
plt.title('Reservation status')     #Title of the figure
plt.bar(['Not cancelled','cancelled'],df['is_canceled'].value_counts(),edgecolor='k',width=0.7)  #setting x axis and y axis lables
plt.show()


#Bar plot is used here because it is categorical data
```

```python
plt.figure(figsize = (8,4))
ax1=sns.countplot(x ='hotel',hue='is_canceled', data = df,palette='Blues')  #using seaborn comparing the hotels cancellations by using plot
legend_labels,_=ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation status in different hotels',size=20)
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.show()
```

```python
resort_hotel=df[df['hotel']=='Resort Hotel']         #filters the hotel column by the value resort hotel
resort_hotel['is_canceled'].value_counts(normalize=True)     #calculates the percentages of canceled and not canceled from is_canceled column
```

```python
city_hotel=df[df['hotel']=='City Hotel']           #filters the hotel column by the value city hotel
city_hotel['is_canceled'].value_counts(normalize=True)    #calculates the percentages of canceled and not canceled from is_canceled column
```

```python
resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean()  #group by is used to group the data
city_hotel=city_hotel.groupby('reservation_status_date')[['adr']].mean()

```

```python
resort_hotel
```

```python
city_hotel
```

```python
plt.figure(figsize=(18,8))
plt.title('Average daily rates of city hotel and resort hotel',fontsize=20)
plt.plot(resort_hotel.index,resort_hotel['adr'],label='Resort Hotel')   #this line plot the variations of adr of resort hotel
plt.plot(city_hotel.index,city_hotel['adr'],label='City Hotel')     #this line plot the variations of adr of city hotel
plt.legend(fontsize=15)
plt.show()
```

```python
df['month']=df['reservation_status_date'].dt.month    #this line extracts the month from reservation status date
plt.figure(figsize=(16,8))
sns.countplot(x='month',hue='is_canceled',data=df,palette='bright')    #plots the graph by taking month on x axis and is_canceled on y axis
legend_labels,_=ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))                           ###these following lines are written for beautification purpose
plt.title('Reservation status per month',size=20)
plt.xlabel('month')
plt.ylabel('Resrvations per month')
plt.legend(['Not cancelled','cancelled'])
plt.show()

```

```python
plt.figure(figsize=(16,8))
plt.title('ADR per month',fontsize=20)
sns.barplot('month','adr',data=df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index())
plt.show()
```

```python
canceled_data=df[df['is_canceled']==1]     #extracts canceled bookings and stores in df
top_10_countries=canceled_data['country'].value_counts()[:10]  #gives top 10  repeated countries fron the country column
plt.figure(figsize=(5,5))
plt.title('Top 10 countries by cancellation')
plt.pie(top_10_countries,autopct='%.2f',labels=top_10_countries.index)   #plots the piechart for top_10_countries
plt.show()
```

```python
canceled_df_adr=canceled_data.groupby(['reservation_status_date'])[['adr']].mean()
canceled_df_adr.reset_index(inplace=True)
canceled_df_adr.sort_values('reservation_status_date',inplace=True)

not_canceled_data=df[df['is_canceled']==0]
not_canceled_df_adr=not_canceled_data.groupby(['reservation_status_date'])[['adr']].mean()
not_canceled_df_adr.reset_index(inplace=True)
not_canceled_df_adr.sort_values('reservation_status_date',inplace=True)

plt.figure(figsize=(20,7))
plt.title('Average Daily Rate')
plt.plot(not_canceled_df_adr['reservation_status_date'],not_canceled_df_adr['adr'],label='not canceled')
plt.plot(canceled_df_adr['reservation_status_date'],canceled_df_adr['adr'],label='canceled')
plt.show()

```

```python
canceled_df_adr=canceled_df_adr[(canceled_df_adr['reservation_status_date']>'2016')&(canceled_df_adr['reservation_status_date']<'2017-09')]
not_canceled_df_adr=not_canceled_df_adr[(not_canceled_df_adr['reservation_status_date']>'2016')&(not_canceled_df_adr['reservation_status_date']<'2017-09')]
#since there are some inconsistancies in the dataset and the data is not present from 2014 to 2015 so we plot line graph only data above 2015 and below 2017
```
